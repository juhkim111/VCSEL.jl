## Organize data for performing variance component selection

### Create a vector of matrices `V`

Given $y$ is a $n\times 1$ phenotype vector, $X$ is a $n \times p$ covariate matrix, and $G_i$ is $n\times q_i$ genotype matrix for $i$-th gene, $i=1,\ldots, m$, we assume 

$$y \sim \text{Normal}(X\beta, \sigma_1^2 V_1 + \ldots + \sigma_m^2 V_m + \sigma_0^2 I)$$

where $V_i$ is a kernel function for $i$-th gene. 

Here we simply use linear kernel function $V_i = G_iG_i^T$. However, any kernel function could be used to model the relationship among the effect sizes of the variants on the phenotypes. For example,
   1. $V_i = G_iW_iG_i^T$ where $W_i = diag(w_1, \ldots, w_{q_i})$ specifies weights of variants in $i$-th gene: uncorrelated effect sizes and corresponds to SKAT
   2. $V_i = G_iW_i1_{q_i}1_{q_i}^TW_iG_i^T$: correponds to the Burden test 

(Dutta, D., Scott, L., Boehnke, M., & Lee, S. (2019). Multi‐SKAT: General framework to test for rare‐variant association with multiple phenotypes. Genetic epidemiology, 43(1), 4-23.)

We use `mat_geno_imputed` for genotype matrix. Here we create two vectors of matrices: 

* `G`: vector of genotype matrix for each gene ($G_i$)
* `V`: vector of covariance matrix for each gene ($V_i = G_iG_i^T$)

We create $V_i$ for gene $i$ by going through the list of candidate gene names and subsetting SNPs that belong to gene $i$.

```julia
using LinearAlgebra
# no. of candidate genes 
m = length(candidates)
# sample size 
n = size(mat_geno_imputed, 1)
# initialize array 
V = Array{Matrix{Float64}}(undef, m + 1)
G = Array{Matrix{Float64}}(undef, m)
for i in 1:m 
    # find indices of SNPs annotated to be gene i 
    tmp_idx = findall(x -> x == candidates[i], df_ann.GENE_NAME)
    # subset genotype matrix using those indices 
    G[i] = mat_geno_imputed[:, tmp_idx]
    # multiply by its transpose 
    V[i] = G[i] * G[i]'
end 
V[end] = Matrix(I, n, n) 
```

`G` is a $150\times 1$ vector whose elements are $324\times q_i$ matrices ($q_i$ denotes the number of SNPs in gene $i$). $i$-th element in `G` is a genotype matrix (matrix of minor allele counts) of $q_i$ SNPs for $i$-th gene, $i=1,\ldots, 150$. 

```julia
G
```

`V` is a $151\times 1$ vector of 324x324 matrices. $i$-th element in `V` corresponds to $i$-th gene, $i=1,\ldots, 150.$ Last element in `V` equals to identity matrix $I_n$, where $n=324$. 

```julia
V
```

## Perform variance component selection 

We assume that response vector $y$ is distributed as 

$$y \sim \text{Normal}(X\beta, \sigma_1^2 V_1 + \cdots + \sigma_{150}^2V_{150} + \sigma_0^2 I_n)$$

where each $\sigma_i^2$ indicates variance component for $i$-th gene, $i=1,\ldots, 150.$ Define $\Omega = \text{Var}(y) = \sigma_1^2 V_1 + \cdots + \sigma_{150}^2V_{150} + \sigma_0^2 I_n.$

Suppose we want to select genes that are associated with response $y.$ This is equivalent to finding non-zero $\sigma_i^2, i=1,\ldots, 150$. This is achieved by minimizing negative log-likelihood function plus penalties:

<a id='objective-function'></a>

$$\frac 12 \log\det\Omega + \frac 12 y^T\Omega^{-1}y + \sum_{i=1}^m P_{\lambda}(\sigma_i) \qquad \qquad (1)$$
    
where $\sum_{i=1}^m P_{\lambda}(\sigma_i)$ is a penalty term that imposes sparsity on variance components for a given tuning parameter $\lambda.$


In this section, we perform variance component selection using `VarianceComponentSelect` package in Julia. First, add `VarianceComponentSelect.jl` package first. 

The package has not yet been registered and must be installed using the repository location. Start Julia and use the `]` key to switch to the package manager REPL

Load the package.

```julia
using VarianceComponentSelect
```

### VCModel

`VCModel` is the fundamental type for variance component model.

### Construct VCModel 

`VCModel` can be initialized by 

where 

* `Y`: $n$-by-$d$ responses 
* `X`: $n$-by-$p$ covariate matrix (if exists)
* `V=(V[1],...,V[m],I)`: a vector of $n$-by-$n$ covariance matrices. The last covariance matrix must be positive definite and usually is the identity matrix 
* `Σ=(Σ[1],...,Σ[m])`: a vector of initial estimates for $d$-by-$d$ variance component parameters. When $d=1$, `Σ` is a $m$-by-$1$ vector. If not supplied, it is set to be a vector of $d$-by-$d$ identity matrix by default. 

```julia
vcm = VCModel(y, X, V);
```

### Estimate variance components without penalty

```julia
vcm_nopenalty = deepcopy(vcm)
vcm_nopenalty, final_obj, niters, = vcselect!(vcm_nopenalty);
```

Output of `vcselect!` contains 

* final objective value, i.e. [negative log-likelihood plus penalty](#objective-function)

```julia
final_obj
```

* fitted model, some fields of which are 
    - `β`: $n$-by-$p$ mean regression coefficients
    - `Σ=(Σ[1],...,Σ[m])`: variance component parameter estimates 

```julia
vcm_nopenalty.β
```

```julia
vcm_nopenalty.Σ 
```

<a id='penwt'></a>
For near future use, we save reciprocal of square root of the estimates as penalty weights. 

```julia
penwt = zeros(m + 1)
penwt[1:m] = 1 ./ sqrt.(vcm_nopenalty.Σ[1:m])
```

* Number of iterations to convergence

```julia
niters
```

### Select variance components using penalty

#### L1 penalty

Now suppose we want to select variance components at $\lambda=2.5$ using lasso, or L1, penalty. That is to say, in the notation of [(1)](#objective-function), $P_{\lambda}(\sigma_j) = \sigma_j$ (we omit absolute value since $\sigma_j \geq 0 \; \forall j$).  To do so, we utilize the following keyword arguments:

* `penfun`: penalty function (e.g. `NoPenalty()`, `L1Penalty()`, `MCPPenalty()`). Default is no penalty `NoPenalty()`.
    - Note: `MCPPenalty()` is only available when response `Y` is a vector, i.e. $d=1$.
* `λ`: tuning parameter value.

```julia
vcm_L1pen = deepcopy(vcm)
vcselect!(vcm_L1pen; λ=2.5, penfun=L1Penalty());
```

Estimates for $\sigma^2=(\sigma_1^2,\ldots, \sigma_{150}^2, \sigma_0^2).$ Note that the last element is an estimate for residual variance component $\sigma_0^2.$

```julia
vcm_L1pen.Σ
```

Selected variance components (i.e. positive ones) are as follows:

```julia
findall(x -> x > 0, vcm_L1pen.Σ[1:end-1])
```

Estimates for fixed effects parameter:

```julia
vcm_L1pen.β
```

Estimated covariance matrix: 

```julia
vcm_L1pen.Ωest
```

#### Adaptive lasso penalty 

Here we use adaptive lasso penalty. 
In the notation of [(1)](#objective-function), $P_{\lambda}(\sigma_j) = \lambda w_j \sigma_j$ where $w_j$ is weight on $\sigma_j.$ To impose weight, we use keyword argument

* `penwt`: $(m+1)$-by-$1$ vector of weights. We do not impose weight on the last variance component $\sigma_0^2.$  Default is [1,1,...,1,0].

We use weight from [earlier section](#penwt). 

```julia
vcm_adL1pen = deepcopy(vcm)
vcselect!(vcm_adL1pen; λ=0.05, penfun=L1Penalty(), penwt=penwt);
```

```julia
vcm_adL1pen.Σ
```

Selected variance components are: 

```julia
findall(x -> x > 0, vcm_adL1pen.Σ[1:end-1])
```

### Obtain solution path 

Now we use `vcselectpath!` function to compute regularization path for a given penalty at a grid of the regularization parameter lambda values. `vcselectpath!` provides options for users to customize, including

* `nλ`: number of `λ` values in the sequence. Default is 100. 
* `λpath`: user-provided sequence of `λ` values in ascending order. Typically the program computes its own `λ` sequence based on `nλ`, but supplying `λpath` overrides this.

#### L1 penalty

```julia
vcmpath_L1 = deepcopy(vcm)
Σ̂path_L1, β̂path_L1, λpath_L1, objpath_L1, niterspath_L1 = vcselectpath!(vcmpath_L1; 
    penfun=L1Penalty())
```

Output of `vcselectpath!` includes

* solution path for variance components: each column represents variance component estimates at certain `λ`. In other words, element in `i`-th row in `j`-th column indicates `i`-th variance component $\sigma_i^2$ at `j`-th `λ` in `λpath`. Note that the last row is for residual variance component, or $\sigma_0^2$.

```julia
Σ̂path_L1
```

* solution path for fixed effects parameter: each column represents fixed effects parameter estimates at certain `λ`. In other words, `j`-th column indicates estimates at `j`-th `λ` in `λpath`.

```julia
β̂path_L1
```

* `λ` values used in the path: by default, a grid of 100 `λ` values is generated internally.

```julia
λpath_L1
```

* final objective values at each `λ`

```julia
objpath_L1
```

* Number of iterations to convergence at each `λ`

```julia
niterspath_L1
```

#### MCP penalty

Here we use MCP penalty and provide a grid of 5 `λ` values.

### Visualize/summarize solution path 

Recall `Σ̂path_L1`:

```julia
Σ̂path_L1
```

We can visualize the path by executing `plotsolpath` function, which requires 

* solution path obtained from `vcselectpath!` 
* `λpath` corresponding to the solution path above. 

By default, `plotsolpath` function produces a legend with rankings of all variance components. By ranking, we mean the order in which variance component enters the path. 

Also note that the effect size of residual variance component (indicated by blue line that stands out) is too large compared to the rest. 

```julia
plotsolpath(Σ̂path_L1, λpath_L1)
```

As the above plot is not informative, let us tweak it a little to make it more informative. We will exclude residual variance component from the solution path and indicate `resvarcomp=false`. We also limit the number of ranks on the legend to 15, or `nranks=15`.

```julia
plotsolpath(Σ̂path_L1[1:end-1, :], λpath_L1; nranks=15, resvarcomp=false)
```

Similarly with solution path for mcp: 

```julia
plotsolpath(Σ̂path_mcp[1:end-1, :], λpath_mcp; resvarcomp=false, nranks=15)
```
